```python
# 5. Створення двох випадкових одновимірних масивів розміром 5
array1 = np.random.randint(1, 11, 5)
array2 = np.random.randint(1, 11, 5)

print(array1)
print(array2)

# Поелементне додавання
addition_result = array1 + array2
print("\nПоелементне додавання:", addition_result)

# Поелементне віднімання
subtraction_result = array1 - array2
print("Поелементне віднімання:", subtraction_result)

# Поелементне множення
multiplication_result = array1 * array2
print("Поелементне множення:", multiplication_result)
```

```python
# 7. Створення двох матриць розміром 2x2 та 2x3 з випадковими цілими числами в діапазоні від 1 до 10
matrix1 = np.random.randint(1, 11, (2, 2))
matrix2 = np.random.randint(1, 11, (2, 3))

print(matrix1)
print(matrix2)

# Перемноження матриць
matrix_product = np.dot(matrix1, matrix2)

print("\nРезультат множення двох матриць:")
print(matrix_product)
```

```python
# 8. Створення матриці розміром 3x3 з випадковими цілими числами в діапазоні від 1 до 10
matrix = np.random.randint(1, 11, (3, 3))

print(matrix)

# Перевірка, чи матриця є квадратною та невиродженою
if np.linalg.det(matrix) != 0:
    # Знаходження оберненої матриці
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
else:
    print("Матриця є виродженою, обернена матриця не існує.")
```

```python
# 9. Створення матриці розміром 4x4 з випадковими дійсними числами в діапазоні від 0 до 1
matrix = np.random.rand(4, 4)

print(matrix)

# Транспонування матриці
transposed_matrix = np.transpose(matrix)

print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
# 10. Створення матриці розміром 3x4 та вектора розміром 4 з випадковими цілими числами в діапазоні від 1 до 10
matrix = np.random.randint(1, 11, (3, 4))
vector = np.random.randint(1, 11, 4)

print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

result = np.dot(matrix, vector)

print("\nРезультат множення:")
print(result)
```

```python
# 16. Створення матриці розміром 3x3 з випадковими дійсними числами в діапазоні від 0 до 1
matrix = np.random.rand(3, 3)

print("Матриця:")
print(matrix)

# Знаходження вектора-стовпчика, що містить суму елементів кожного рядка матриці
sum_of_rows = np.sum(matrix, axis=1)

print("\nСума елементів кожного рядка матриці:")
print(sum_of_rows)
```

```python
# 17. Створення матриці розміром 3x4 з довільними цілими числами
matrix = np.random.randint(1, 11, (3, 4))

print("Початкова матриця:")
print(matrix)

# Створення матриці з квадратами елементів першої матриці
squared_matrix = np.square(matrix)

print("\nМатриця з квадратами чисел:")
print(squared_matrix)
```
